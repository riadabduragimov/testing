```python
import re
```

## 1
Which matches a word containing vowels in it (a,ı,o,u,e,ə,i,ö,ü).

```python
text="""Qəmu bimarinə canan dəvayi-dərd edər ehsan.
Neçin qılmaz mənə dərman, məni bimar sanmazmı? tmm slm xlxs"""
re.findall(r'\w+[aıoueəiöü]\w+',text)
```

## 2
Which matches a word containing vowels not at the start or end of the word (a,ı,o,u,e,ə,i,ö,ü).

```python
text="""Qəmu bimarinə canan dəvayi-dərd edər ehsan.
Neçin qılmaz mənə dərman, məni bimar sanmazmı? tmm slm xlxs"""
for i in text.split():
    word=re.findall(r'(^[^aıoueəiöü][\w]+[^aıoueəiöü?]$)',i)
    if word==[]:
        continue
    else: print(word)
```

```python
text="""Qəmu bimarinə canan dəvayi-dərd edər ehsan.
Neçin qılmaz mənə dərman, məni bimar sanmazmı? Ələk tmm slm xlxs"""

re.findall(r'\b[^aıoueəiöüAIOUEƏİÖÜ\s]\w*[aıoueəiöü]+\w*[^aıoueəiöü\s]\b',text)
```

## 3
Checks for a number at the end of a given string

```python
text3='mnkmfrefg sacsd33'
text4='mnkmfrefg sacsd'
print(re.findall('.*\d+$',text3))
print(re.findall('.*\d+$',text4))
```

## 4
Which matches a word at the end of string, with optional question mark and exclamation point.

```python
print(re.findall(r'\w+[?!]{1,2}\Z','SAlam!'))
print(re.findall(r'\w+[?!]{1,2}\Z','Python Best	Programming   Language for me?'))
print(re.findall(r'\w+[?!]{1,2}\Z','Python Best	Programming   Language for me!?'))
```

## 5
Replaces whitespaces with a tab and vice versa.

```python
text1='Python Best Programming    Language for me'
print(text1)
print(re.sub(r'\s',r'\t',text1))
print(re.sub(r'\s{4}',' ',text1))
```
