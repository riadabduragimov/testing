## 9
Get the dates of yesterday, today and tomorrow using np.datetime64() and np.timedelta64().

```python
today= np.datetime64('Today','D')
yesterday=np.datetime64('Today','D') -np.timedelta64(1,'D')
tomorrow=np.datetime64('Today') +np.timedelta64(1,'D')
yesterday,today,tomorrow
```

# Visualization

```python
cars=pd.read_excel('cars.xlsx')
cars.head()
```

## 1. 
Show relationship between **highway** and **city** miles per gallon.

a. Use scatter in matplotlib.

b. Interpret how are variables correlated according to scatter.

```python
plt.scatter(x=cars['hwy'],y=cars['cty'])
plt.show()
```

## OR

```python
sns.scatterplot(x='hwy',y='cty',data=cars)
plt.show()
```

## 2. 
Show distributions and scatters between all variables. While hue equals to type of

Find according to scatterplot.

```python
cars.head()
```

```python
sns.pairplot(data=cars,hue='drv')
```

```python
sns.scatterplot(x='hwy',y='cty',data=cars,hue='drv')
plt.show()
```

## 3
Which type of car is most frequent in dataset? Show by using countplot.

```python
sns.countplot(x='class',data=cars)
plt.show()
```

## 4
Display number of cylinders for each drive train in bar chart. Which drive train is

```python
cars.head()
```

```python
cars['drv'].unique()
```
